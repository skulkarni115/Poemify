```python
import json
import os
import requests
from google.colab import drive
import shutil

# Mount Google Drive
drive.mount('/content/drive')

def download_images_from_json(json_file_path, download_folder):
    # Create download folder if it doesn't exist
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)

    # Load JSON data
    with open(json_file_path, 'r') as file:
        data = json.load(file)

    # Iterate through each dictionary in the JSON data
    for item in data:
        image_url = item.get('image_url')
        if image_url:
            # Extract image filename from the URL
            filename = os.path.join(download_folder, f"{item['id']}.jpg")

            # Download image
            response = requests.get(image_url)
            if response.status_code == 200:
                with open(filename, 'wb') as image_file:
                    image_file.write(response.content)
            else:
                print(f"Failed to download image for ID {item['id']}")

json_file_path = '/content/sample_data/multim_poem.json'
download_folder = '/content/drive/MyDrive/Colab Notebooks/images'
download_images_from_json(json_file_path, download_folder)
```
